# EX 1.2

Survival distribution

```
print("Survival rate:")
print(df['survived'].value_counts(normalize=True).round(3) * 100)

sns.countplot(x='survived', data=df, palette='viridis')
plt.title('Survival Count')
plt.show()
```

Survival by pclass

```
print("\nSurvival rate by pclass:")
print(df.groupby('pclass')['survived'].mean().round(3) * 100)

sns.countplot(x='pclass', hue='survived', data=df, palette='viridis')
plt.title('Survival by Passenger Class')
plt.show()
```

Survival by sex

```
print("\nSurvival rate by sex:")
print(df.groupby('sex')['survived'].mean().round(3) * 100)

sns.countplot(x='sex', hue='survived', data=df, palette='viridis')
plt.title('Survival by Sex (0 = Male, 1 = Female)')
plt.show()
```

Age distribution

```
plt.figure(figsize=(10,5))
sns.histplot(df['age'].dropna(), kde=True, bins=30)
plt.title('Age Distribution')
plt.show()

sns.boxplot(x='survived', y='age', data=df, palette='viridis')
plt.title('Age by Survival')
plt.show()
```

Fare by pclass

```
sns.boxplot(x='pclass', y='fare', data=df, palette='viridis')
plt.title('Fare by Passenger Class')
plt.yscale('log')
plt.show()
```

Correlation matrix (numeric columns only)

```
numeric_cols = ['pclass', 'survived', 'sex', 'age', 'sibsp', 'parch', 'fare']
plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
